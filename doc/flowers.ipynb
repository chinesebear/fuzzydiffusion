```python
import os
from PIL import Image
import numpy as np
from tqdm import tqdm
path = '/home/yang/sda/github/fuzzydiffusion/output/datasets/nelorth/oxford-flowers/train/'
files = os.listdir(path)
record = {}
for filename in tqdm(files):
    if filename.find('.jpg') == -1:
        continue
    img_path = path + filename
    img = Image.open(img_path)
    img = np.array(img)
    H, W, C = img.shape
    img_h_w = str(H)+'*'+str(W)
    if img_h_w in record.keys():
        record[img_h_w] = record[img_h_w] + 1
    else:
        record[img_h_w] = 1
print(record)
```
