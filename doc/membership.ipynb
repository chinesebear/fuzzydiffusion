```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
import pandas as pd
import numpy as np
```

```python
# Label=["landspace","animal","human"]
# Color=["lightsteelblue","cornflowerblue","royalblue"]

df=pd.read_csv("membershipcurve.csv")
landspace=df['landspace'].values[:50]
animal=df['animal'].values[:50]
human=df['human'].values[:50]
xlen=len(landspace)

fig, axe = plt.subplots(nrows=1, ncols=2, figsize=(8, 4),dpi=300)

plt.subplot(1, 2, 1)
plt.scatter(range(xlen),landspace,c="skyblue",marker='o',s=20, label='landspace')
plt.scatter(range(xlen),animal,c="darkorange",marker='o',s=20, label='animal')
plt.scatter(range(xlen),human,c="slategrey",marker='o',s=20, label='human')
plt.plot(range(xlen), 0.33*np.ones(xlen),c="red",linestyle='--',linewidth=1.0)
plt.xlabel("Diffusion Steps")
plt.ylabel("Membership Degree")
plt.grid()
plt.legend()

plt.subplot(1, 2, 2)
plt.bar(range(xlen),landspace, color="skyblue", label='landspace', alpha=0.6, width=0.6)
plt.bar(range(xlen), animal, color="darkorange", label='animal', alpha=0.6, width=0.6, bottom=landspace)
plt.bar(range(xlen), human, color="slategrey", label='human', alpha=0.6, width=0.6, bottom=landspace+animal)
plt.xlabel("Diffusion Steps")
plt.ylabel("Membership Distrubition")
# plt.grid()
# plt.legend()

plt.rcParams.update({'font.size': 15})
fig.tight_layout()#
plt.savefig("figure.membership.svg", format = "svg")
plt.savefig("figure.membership.jpg", format = "jpg")
plt.show()
```
