```python
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
import pandas as pd
```

```python
df = pd.read_csv("stable.csv")
```

```python
epoch = df['Epoch']
fdm = df['FDM'] #DFS
md = df['MD']
sdg = df['SDG']
ldm = df['LDM']

fig = plt.figure(figsize=(10, 6), dpi=300)

plt.plot(epoch, fdm, label="DFS", color='red')
plt.plot(epoch, md, label="MD", color='tan')
plt.plot(epoch, sdg, label="SDG", color='skyblue')
plt.plot(epoch, ldm, label="LDM", color='slategrey')
plt.title("Training")
plt.xlabel("Epoch")
plt.ylabel("FID")
plt.legend()
plt.grid()

plt.rcParams.update({'font.size': 20})
plt.savefig("figure.stability.svg", format = "svg",bbox_inches='tight')
plt.savefig("figure.stability.jpg", format = "jpg",bbox_inches='tight')
plt.show()
```
