```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
import numpy as np
```

```python
# 生成一组随机数据
data = [np.random.randn(1000)]
dict = {"truck":7513,
"tree":7036,
"road":10230,
"person":24736,
"teddy":5037,
"raincoat":33,
"boots":209,
"horse":8401,
"grass":13010,
"sign":14045,
"fuzzy diffusion": 0}
prompts = list(dict.keys())
counts = list(dict.values())
total = len(prompts)
# print(prompts)
# print(counts)

plt.barh(range(total), counts, height=0.7, color='skyblue', alpha=0.8)      # 从下往上画
plt.yticks(range(total), prompts)
plt.xlim(right = 50000)
plt.xlabel("Prompt Count")
plt.title("Prompt of MS_COCO dataset")
for x, y in enumerate(counts):
    freq = format(y * 100 / 591753, '.2f')
    plt.text(y + 0.2, x - 0.1, '%s%%' % freq)
plt.savefig("prompt_freq.svg", format = "svg")
plt.savefig("prompt_freq.jpg", format = "jpg")
# 显示图表
plt.show()
```
