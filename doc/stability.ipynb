```python
import matplotlib.font_manager as fm
fm.fontManager.addfont('/usr/share/fonts/truetype/times.ttf')
import matplotlib.pyplot as plt
plt.rc('font',family='Times New Roman')
import pandas as pd
```

```python
df = pd.read_csv("/home/yang/sda/github/fuzzydiffusion/doc/stable.csv")
```

```python
epoch = df['Epoch']
fdm = df['FDM']
md = df['MD']
sdg = df['SDG']
ldm = df['LDM']

plt.plot(epoch, fdm, label="FDM", color='red')
plt.plot(epoch, md, label="MD", color='tan')
plt.plot(epoch, sdg, label="SDG", color='skyblue')
plt.plot(epoch, ldm, label="LDM", color='slategrey')
plt.title("Training")
plt.xlabel("Epoch")
plt.ylabel("FID")
plt.legend()
plt.savefig("stability.svg", format = "svg")
plt.savefig("stability.jpg", format = "jpg")

plt.tight_layout(pad=0)#调整整体空白
plt.show()
```
